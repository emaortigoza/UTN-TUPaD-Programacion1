# **Enunciado**
Se desea desarrollar un programa que registre personas en un archivo llamado personas.csv. Cada persona debe tener los siguientes datos: nombre, edad y nacionalidad. El programa debe comenzar creando el archivo y escribiendo un primer registro con datos ingresados por el usuario. Luego, debe permitir agregar ms personas de a una por vez, hasta que el usuario decida terminar. Todos los registros deben guardarse en el archivo separados por comas, siguiendo el formato: nombre,edad,nacionalidad. Al finalizar, el programa debe mostrar todos los datos almacenados en el archivo.


---


### **Detalles**
  **Objetivo:** Desarrollar un programa que registre personas en un archivo CSV llamado "personas.csv".

  **Datos por persona:** nombre, edad y nacionalidad

  **Funcionalidades:** Crear archivo inicial, agregar personas una por vez, mostrar todos los registros al final.


---


### **Estructura**
**Crear archivo inicial:** El programa debe crear "personas.csv" y escribir el primer registro con datos ingresados por el usuario.

**Agregar ms personas:** Permitir al usuario agregar ms personas de a una por vez hasta que decida terminar.

**Mostrar resultados:** Al finalizar, leer y mostrar todos los datos almacenados en el archivo con formato legible.

```python
nombre = input("Ingres el nombre: ")
edad = input("Ingres la edad: ")
nacionalidad = input("Ingres la nacionalidad: ")

with open("personas.csv", "w") as archivo:
    archivo.write(f"{nombre},{edad},{nacionalidad}\n")

while True:
    respuesta = input("Quers agregar otra persona? (s/no): ")
    if respuesta.lower() != "si":
        break

    nombre = input("Ingres el nombre: ")
    edad = input("Ingres la edad: ")
    nacionalidad = input("Ingres la nacionalidad: ")

    with open("personas.csv", "a") as archivo:
        archivo.write(f"{nombre},{edad},{nacionalidad}\n")

print("\nPersonas registradas en el archivo:")
with open("personas.csv", "r") as archivo:
    for linea in archivo:
        datos = linea.strip().split(",")
        print(f"Nombre: {datos[0]}, Edad: {datos[1]}, Nacionalidad: {datos[2]}")
```
